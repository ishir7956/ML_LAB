# Q1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, log_loss

data = pd.read_csv('/home/student/Desktop/210962062/WEEK_7/lab7.csv')
X1 = np.array(data['X1'].values)
X2 = np.array(data['X2'].values)
Y = np.array(data['Y'].values)

# MANUAL PROGRAM
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

B0 = 0
B1 = 0
B2 = 0
learning_rate = 0.01
n_iterations = 1000

errors = []
B1_values = []
B2_values = []

for _ in range(n_iterations):
    Z = B0 + B1 * X1 + B2 * X2
    predictions = sigmoid(Z)
    error = -np.mean(Y * np.log(predictions) + (1 - Y) * np.log(1 - predictions))
    errors.append(error)
    B1_values.append(B1)
    B2_values.append(B2)
    B0 -= learning_rate * np.mean(predictions - Y)
    B1 -= learning_rate * np.mean((predictions - Y) * X1)
    B2 -= learning_rate * np.mean((predictions - Y) * X2)

predicted_labels = (B0 + B1 * X1 + B2 * X2) > 0.5
accuracy = np.mean(predicted_labels == Y)

print("Final Coefficient B0:", B0)
print("Final Coefficient B1:", B1)
print("Final Coefficient B2:", B2)
print("Accuracy:", accuracy)

for i in range(len(errors)):
    print(f"Iteration {i + 1}: Error = {errors[i]}, B1 = {B1_values[i]}, B2 = {B2_values[i]}")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(B1_values, errors)
plt.xlabel('B1')
plt.ylabel('Error')
plt.title('B1 vs. Error')

plt.subplot(1, 2, 2)
plt.scatter(B2_values, errors)
plt.xlabel('B2')
plt.ylabel('Error')
plt.title('B2 vs. Error')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('/home/student/Desktop/210962062/WEEK_7/lab7.csv')

X1 = np.array(df['X1'].values)
Y = np.array(df['Y'].values)

# MANUAL PROGRAM
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

B0 = 0
B1 = 0
learning_rate = 0.01
n_iterations = 1000

errors = []
B1_values = []

for _ in range(n_iterations):
    Z = B0 + B1 * X1
    predictions = sigmoid(Z)
    error = -np.mean(Y * np.log(predictions) + (1 - Y) * np.log(1 - predictions))
    errors.append(error)
    B1_values.append(B1)
    B0 -= learning_rate * np.mean(predictions - Y)
    B1 -= learning_rate * np.mean((predictions - Y) * X1)

predicted_labels = (B0 + B1 * X1) > 0.5
accuracy = np.mean(predicted_labels == Y)

print("Final Coefficient B0:", B0)
print("Final Coefficient B1:", B1)
print("Accuracy:", accuracy)

for i in range(len(errors)):
    print(f"Iteration {i + 1}: Error = {errors[i]}, B1 = {B1_values[i]}")

# Plot the graph of B1 vs. Error
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(B1_values, errors)
plt.xlabel('B1')
plt.ylabel('Error')
plt.title('B1 vs. Error')

# Visualize Binary Cross-Entropy Function
# Generate X values for the plot
X_plot = np.linspace(min(X1), max(X1), 100)
# Calculate the logistic function for Y=0 and Y=1 separately
predictions_y0 = 1 / (1 + np.exp(-(B0 + B1 * X_plot)))  # Y=0
predictions_y1 = 1 - predictions_y0  # Y=1
# Calculate the cost for Y=0 and Y=1 separately
cost_y0 = -np.log(1 - predictions_y0)
cost_y1 = -np.log(predictions_y1)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(X_plot, cost_y0, label='Y=0')
plt.xlabel('X')
plt.ylabel('Cost')
plt.title('Cost for Y=0')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(X_plot, cost_y1, label='Y=1')
plt.xlabel('X')
plt.ylabel('Cost')
plt.title('Cost for Y=1')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#MANUAL PROGRAM
X1 = df['X1']
X2 = df['X2']
Y = df['Y']

# Initialize coefficients (B0, B1, and B2) with random values
np.random.seed(0)
B0 = np.random.rand()
B1 = np.random.rand()
B2 = np.random.rand()

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Set hyperparameters
learning_rate = 0.01
epochs = 5
total_iterations = 50

# Lists to store accuracy, epoch, and coefficient values
accuracy_values = []
epoch_values = []
B0_values = []
B1_values = []
B2_values = []
error_values = []

# Perform SGD
for epoch in range(epochs):
    for iteration in range(total_iterations):
        # Calculate predicted values
        Z = B0 + B1 * X1 + B2 * X2
        Y_pred = sigmoid(Z)

        # Calculate error
        error = -1 / len(Y) * np.sum(Y * np.log(Y_pred) + (1 - Y) * np.log(1 - Y_pred))

        # Update coefficients using gradient descent
        dB0 = 1 / len(Y) * np.sum(Y_pred - Y)
        dB1 = 1 / len(Y) * np.sum((Y_pred - Y) * X1)
        dB2 = 1 / len(Y) * np.sum((Y_pred - Y) * X2)

        B0 -= learning_rate * dB0
        B1 -= learning_rate * dB1
        B2 -= learning_rate * dB2

        # Calculate accuracy for this iteration
        Y_pred_binary = [1 if p >= 0.5 else 0 for p in Y_pred]
        accuracy = np.mean(Y_pred_binary == Y)

        # Append accuracy, epoch, and coefficient values
        accuracy_values.append(accuracy)
        epoch_values.append(epoch * total_iterations + iteration)
        B0_values.append(B0)
        B1_values.append(B1)
        B2_values.append(B2)
        error_values.append(error)

# Plot the graph of epoch vs. accuracy
plt.plot(epoch_values, accuracy_values)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Epoch vs. Accuracy')
plt.show()

# Print the values of B0, B1, and B2 for every iteration
for i in range(len(epoch_values)):
    print(f"Iteration {epoch_values[i]} - B0: {B0_values[i]}, B1: {B1_values[i]}, B2: {B2_values[i]}, Error: {error_values[i]}")
```
